```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
s_in_d = 86400
d10_in_sec = 10 * s_in_d

dt=120
d10_in_dt = d10_in_sec/dt

dts = np.arange(d10_in_dt)

dt_relax_s = 91440
dt_relax = 91440/dt
```

```python
utau0 = 0.8

cosonly = np.zeros_like(dts)
for kt in dts:
    cosonly[int(kt)] = utau0 * 0.5*(1 - np.cos((kt*dt*0.0001373)/2))
```

```python
fig, ax = plt.subplots(1,1,figsize=(20, 5))
ax.plot(dts, cosonly)
ax.axvline(dt_relax)
ax.grid()
```

```python
utau1 = 0.1

tanhonly = np.zeros_like(dts)
for kt in dts:
    tanhonly[int(kt)] = ( np.tanh( ( kt - ( dt_relax/2 ) ) / ( dt_relax/6 ) ) + 1 ) * (utau1/2)
```

```python
fig, ax = plt.subplots(1,1,figsize=(20, 5))
ax.plot(dts, tanhonly)
ax.axvline(dt_relax)
ax.grid()
```
