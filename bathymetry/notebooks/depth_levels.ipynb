Basically, ppdzmin is the depth interval it aspires to have. So, hypothetically, if you had no stretching or squeezing, you could do full depth divided the depth intervals and give that as the value of ppdzmin and that should be the more or less the depth interval throughout. 

So, what I'm seeing is that it tries to maintain ppdzmin but according to the ppkth, it will try to reach pphmax by the final depth interval, so it will use ppacr to guide how much it needs to stretch. It's a bit confusing, but I think of it this way: the higher the ppacr, the deeper you'll find the ppkth depth level. This extends to saying, that the higher the ppacr, the more spaced out the depth levels will be before ppkth. So, I'll rewrite my namelist comment for ppacr.

Previously, I thought that the value for e3w_1d #1 was the depth interval between depths gdepw_1d #1 and #2. However, now I believe the other way around. e3w_1d #20 corresponds to gdepw_1d #20 and #19. The significance of ppkth is difficult to see in most cases except, test 11. In this case, the depth intervals remain constant at the ppdzmin value up until e3w_1d #14 and then change at e3w_1d #15. The interval e3w_1d #15 corresponds to gdepw_1d #15 and #14. The given value for ppkth was 16. So for now I interpret ppkth as either the first or second depth interval where the profile tries stretch to reach pphmax. I'll rewrite the namelist comment for this too.
