Basically, ppdzmin is the depth interval it aspires to have. So, say you have zero stretching, you could do 1300/20 and give that as the value of ppdzmin and that should be the interval throughout (try this). However, it will try to keep it that way for a while until it has to start stretching in order to reach 1300 by the last (20th) level. I think that below the ppkth value, it just goes to overdrive to try to reach the 1300. I'm looking at values for e3w to visualize.
