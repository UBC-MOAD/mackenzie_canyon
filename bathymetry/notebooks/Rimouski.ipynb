```python
import numpy as np
import netCDF4 as nc
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import cmocean as cmo
%matplotlib inline
```

```python
import sys
sys.path.append('/ocean/imachuca/Canyons/mackenzie_canyon/tools/')
import functions_plot
(cmap, norm) = functions_plot.Colormap()
from matplotlib.colors import ListedColormap
cmap_r = ListedColormap(cmap.colors[::-1])
```

```python
ideal_file = nc.Dataset('/ocean/imachuca/Canyons/mackenzie_canyon/bathymetry/NEMO_files/idealized/ideal_bathy_05.nc')
real_file = nc.Dataset('/ocean/imachuca/Canyons/mackenzie_canyon/bathymetry/NEMO_files/realistic/real_bathy_02.nc')
ideal = ideal_file['Bathymetry']
real = real_file['Bathymetry']
```

```python
levels = np.linspace(0 ,1300, 50)
lw = 2
ticks = np.linspace(0 ,1300, 5)
```

```python
fig, ax = plt.subplots(1,1,figsize=(14, 8))
ax.pcolormesh(ideal, cmap=cm.Blues)
c = ax.contour(ideal[:], levels = levels, cmap = cmap_r, linewidths=lw)
ax.set_title('Idealized Bathymetry of Mackenzie Canyon', fontsize=16)
ax.set_xlabel('Alongshore (x) indices', fontsize=14)
ax.set_ylabel('Cross-shore (y) indices', fontsize=14)
ax.text(180, 160, 'Arctic Ocean', fontsize=15, color='white', fontweight='bold')
ax.text(180, 15, 'Mackenzie Shelf', fontsize=15, color='black', fontweight='bold')
cb = fig.colorbar(c, ticks=ticks)
cb.lines[0].set_linewidth(10)
cb.set_label('Depth [m]', fontsize=13.5)
```

```python
fig, ax = plt.subplots(1,1,figsize=(14, 8))
ax.pcolormesh(real, cmap=cm.Blues)
c = ax.contour(real[:], levels = levels, cmap = cmap_r, linewidths=lw)
ax.set_title('Realistic Bathymetry of Mackenzie Canyon', fontsize=16)
ax.set_xlabel('Alongshore (x) indices', fontsize=14)
ax.set_ylabel('Cross-shore (y) indices', fontsize=14)
ax.text(180, 160, 'Arctic Ocean', fontsize=15, color='white', fontweight='bold')
ax.text(180, 15, 'Mackenzie Shelf', fontsize=15, color='black', fontweight='bold')
cb = fig.colorbar(c, ticks = ticks)
cb.lines[0].set_linewidth(10)
cb.set_label('Depth [m]', fontsize=13.5)
```
